This notebook compares the spatial (Δx) and temporal (Δt) distributions of three models:
- SGCV (Superluminal Graviton Condensate Vacuum)
- ALP (Axion-Like Particle)
- KM (Kinetic Mixing Dark Photon)

We generate:
- A scatter plot of Δx vs Δt
- Comparative histograms for Δx and Δt


```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)  # reproducibility
```

```python
# Generate mock data for SGCV, ALP, and KM models
data = {}
models = ['SGCV', 'ALP', 'KM']
params = {
    'SGCV': {'dx_mean': 0.2, 'dt_mean': 0.1, 'dx_std': 0.05, 'dt_std': 0.03},
    'ALP': {'dx_mean': 0.1, 'dt_mean': 0.2, 'dx_std': 0.02, 'dt_std': 0.05},
    'KM': {'dx_mean': 0.15, 'dt_mean': 0.15, 'dx_std': 0.03, 'dt_std': 0.03},
}
N = 1000
for model in models:
    p = params[model]
    dx = np.random.normal(p['dx_mean'], p['dx_std'], N)
    dt = np.random.normal(p['dt_mean'], p['dt_std'], N)
    data[model] = {'dx': dx, 'dt': dt}
```

## Δx vs Δt Plot
Each point represents a particle event for the corresponding model.

```python
# Δx vs Δt Scatter Plot
plt.figure(figsize=(10, 6))
for model in models:
    plt.scatter(data[model]['dx'], data[model]['dt'], label=model, alpha=0.6)
plt.xlabel('Δx')
plt.ylabel('Δt')
plt.title('Δx vs Δt Comparison')
plt.legend()
plt.grid(True)
plt.show()
```

## Histogram of Δx
Comparison of spatial resolution between models.

```python
# Histogram of Δx
plt.figure(figsize=(10, 6))
for model in models:
    plt.hist(data[model]['dx'], bins=50, alpha=0.6, label=model, density=True)
plt.xlabel('Δx')
plt.ylabel('Probability Density')
plt.title('Δx Histogram Comparison')
plt.legend()
plt.grid(True)
plt.show()
```

## Histogram of Δt
Comparison of time-of-flight resolution between models.

```python
# Histogram of Δt
plt.figure(figsize=(10, 6))
for model in models:
    plt.hist(data[model]['dt'], bins=50, alpha=0.6, label=model, density=True)
plt.xlabel('Δt')
plt.ylabel('Probability Density')
plt.title('Δt Histogram Comparison')
plt.legend()
plt.grid(True)
plt.show()
```
